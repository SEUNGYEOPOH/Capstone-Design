## Velocity

```python
su = sub[0:1001]
dange = danger1[0:1001]
japa = japan[0:1001]
japa2 = japan2[0:1001]
bu = bus[0:1001]
```

```python
plt.figure(figsize=(16,8))
plt.plot(bu['Velocity'],
         color='C7',label='Bus Average Velocity')

plt.plot(dange['Velocity'],
         color='C1',label='Danger State Average Velocity')
plt.plot(japa['Velocity'],
         color='C2',label='Kabukicho Average Velocity')
plt.plot(japa2['Velocity'],
         color='C6', label='Shibuya Average Velocity')

plt.plot(su['Velocity'],
         color='C9',label='subway Average Velocity')
plt.ylabel('Velocity')
plt.xlabel('Time')
plt.legend()
plt.show()
```

## Magnitude

```python
plt.figure(figsize=(16,8))
plt.plot(bu['Magnitude'],
         color='C7',label='Bus Average Magnitude')

plt.plot(dange['Magnitude'],
         color='C1',label='Danger State Average Magnitude')
plt.plot(japa['Magnitude'],
         color='C2',label='Kabukicho Average Magnitude')
plt.plot(japa2['Magnitude'],
         color='C6', label='Shibuya Average Magnitude')

plt.plot(su['Magnitude'],
         color='C9',label='subway Average Magnitude')
plt.ylabel('Magnitude')
plt.xlabel('Time')
plt.legend()
plt.show()
```

## Direction

```python
plt.figure(figsize=(16,8))
plt.plot(bu['Direction'],
         color='C7',label='Bus Average Direction')

plt.plot(dange['Direction'],
         color='C1',label='Danger State Average Direction')
plt.plot(japa['Direction'],
         color='C2',label='Kabukicho Average Direction')
plt.plot(japa2['Direction'],
         color='C6', label='Shibuya Average Direction')

plt.plot(su['Direction'],
         color='C9',label='subway Average Direction')
plt.ylabel('Direction')
plt.xlabel('Time')
plt.legend()
plt.show()
```

Velocity(속도)가 크다는 것은 활동성이 크다는 의미입니다. 위험 상황에서는 활동성이 높아지는 경향이 있기 때문에 Velocity의 큰 진폭은 해당 지역에서 군중 유동화가 발생하였음을 시사할 수 있습니다.

군중 유동화는 일반적으로 인구 밀도의 증가와 활동성의 증가로 특징짓는데, 위험 상황에서는 사람들이 더 많은 움직임을 보일 가능성이 높습니다. 예를 들어, 재난 현장, 사건 발생 지역, 대규모 이벤트 장소 등에서는 많은 사람들이 모여 활동하고 이동할 수 있습니다. 이로 인해 Velocity 값이 큰 진폭을 나타낼 수 있습니다.

따라서, Velocity가 크다는 것은 해당 지역에서 군중 유동화가 발생하고 활동성이 크다는 의미로 해석할 수 있습니다. 이는 위험 상황을 감지하고 위험 관리 조치를 취하는 데 도움을 줄 수 있습니다.

```python
plt.figure(figsize=(16,8))
plt.scatter(bu['Velocity'],bu['Direction'],
         color='C7',label='Bus')

plt.scatter(dange['Velocity'],dange['Direction'],
         color='C1',label='Danger State')
plt.scatter(japa['Velocity'],japa['Direction'],
         color='C2',label='Kabukicho')
plt.scatter(japa2['Velocity'],japa2['Direction'],
         color='C6', label='Shibuya')

plt.scatter(su['Velocity'],su['Direction'],
         color='C9',label='subway')
plt.ylabel('Direction')
plt.xlabel('Velocity')
plt.legend()
plt.show()
```
