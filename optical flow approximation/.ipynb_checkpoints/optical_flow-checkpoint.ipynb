```python
import re
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sub = pd.read_csv("./opticalflowdata/subway_optical.csv")
danger = pd.read_csv("./opticalflowdata/danger_opt.csv")
japan = pd.read_csv("./opticalflowdata/japan_opt.csv")
japan2 = pd.read_csv("./opticalflowdata/japan2_opt.csv")
bus = pd.read_csv("./opticalflowdata/bus_optical.csv")
```

```python
danger
```

```python
arr = []
for i in range(len(danger['Direction'])):
    numbers = re.findall(r'\d+\.\d+', danger['Direction'][i])
    total_sum = sum(float(num) for num in numbers)
    num_count = len(numbers)
    mean = total_sum / num_count
    arr.append(mean)
danger['Direction']=arr
```

```python
danger
```

```python
arr = []
for i in range(len(danger['Magnitude'])):
    numbers = re.findall(r'\d+\.\d+', danger['Magnitude'][i])
    total_sum = sum(float(num) for num in numbers)
    num_count = len(numbers)
    if total_sum!=0:    
        mean = total_sum / num_count
    else:
        mean=0
    arr.append(mean)
danger['Magnitude']=arr
```

```python
danger[danger['Magnitude']>0]
```

```python
danger.to_csv("./opticalflowdata/danger_opt.csv")
```

## Velocity

```python
su = sub[0:1001]
dange = danger[0:1001]
japa = japan[0:1001]
japa2 = japan2[0:1001]
bu = bus[0:1001]
```

```python
plt.figure(figsize=(16,8))
plt.plot(bu['Velocity'],
         color='C7',label='Bus Average Velocity')

plt.plot(dange['Velocity'],
         color='C1',label='Danger State Average Velocity')
plt.plot(japa['Velocity'],
         color='C2',label='Kabukicho Average Velocity')
plt.plot(japa2['Velocity'],
         color='C6', label='Shibuya Average Velocity')

plt.plot(su['Velocity'],
         color='C9',label='subway Average Velocity')
plt.ylabel('Velocity')
plt.xlabel('Time')
plt.legend()
plt.show()
```
